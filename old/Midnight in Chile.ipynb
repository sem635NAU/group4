```python
import numpy as np
import pandas as pd
import math
```

```python
obs_df = pd.read_csv('obs_df.csv')
```

```python
obs_JD = obs_df['observationStartMJD'].to_numpy()
```

```python
october = 60949.166666666664
obs_JD = obs_JD[obs_JD>october]
obs_JD
```

```python
save_times = []
start = october
end = 0
for i in range(len(obs_JD[:-1])):
    if abs(obs_JD[i+1]-obs_JD[i]) > 5/24:
        end = obs_JD[i]
        #print(obs_JD[i-10:i+10])
        #print(f'halfway is {(start+end)/2}')
        midnight = (start+end)/2
        start = obs_JD[i+1]
        save_times.append(midnight)

save_times
```

```python
numpy_array = np.array(save_times)
```

```python
df = pd.DataFrame(numpy_array, columns=["Midnight in Chile"])
```

```python
df
```

```python
df.to_csv('midnight_in_Chile_df.csv', index=False)
```
